```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
iris_data['species'] = LE.fit_transform(iris_data['species'])
iris_data.head()
```

```python
X = iris_data.drop(['species'], axis = 1)
Y = iris_data['species']
```

```python
X.values.reshape(-1,1)
Y.values
```

# Splitting the Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,train_size = 0.6, random_state = 42)
```

# Model Training and Evaluation 

## Support Vector Classifier

```python
from sklearn.svm import SVC
SVC_Model = SVC()
SVC_Model.fit(X_train, Y_train)
Y_pred_SVC = SVC_Model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
accuracy_SVC = accuracy_score(Y_test, Y_pred_SVC)
print("Accuracy of the model is", accuracy_SVC*100)
cnf_matrix = confusion_matrix(Y_test, Y_pred_SVC)

sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu", fmt='g')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
labels = ['Iris Setosa','Iris Virginia','Iris Versicolor']
print(classification_report(Y_test, Y_pred_SVC))


```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
LR_Model = LogisticRegression()
LR_Model.fit(X_train, Y_train)
Y_pred_LR = LR_Model.predict(X_test)
```

```python
accuracy_LR = accuracy_score(Y_test, Y_pred_LR)
print("Accuracy of the model is", accuracy_LR*100)
cnf_matrix = confusion_matrix(Y_test, Y_pred_LR)

sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu", fmt='g')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
labels = ['Iris Setosa','Iris Virginia','Iris Versicolor']
print(classification_report(Y_test, Y_pred_LR))

```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
RFC_Model = RandomForestClassifier()
RFC_Model.fit(X_train, Y_train)
Y_pred_RFC = RFC_Model.predict(X_test)
```

```python
accuracy_RFC = accuracy_score(Y_test, Y_pred_RFC)
print("Accuracy of the model is", accuracy_RFC*100)
cnf_matrix = confusion_matrix(Y_test, Y_pred_RFC)

sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu", fmt='g')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
labels = ['Iris Setosa','Iris Virginia','Iris Versicolor']
print(classification_report(Y_test, Y_pred_RFC))


```

## KNeighbours Classifier

```python
from sklearn import neighbors
classifier = neighbors.KNeighborsClassifier()
KNN_Model = classifier.fit(X_train, Y_train)
Y_pred_KNN = KNN_Model.predict(X_test)
```

```python
accuracy_KNN = accuracy_score(Y_test, Y_pred_KNN)
print("Accuracy of the model is", accuracy_KNN*100)
cnf_matrix = confusion_matrix(Y_test, Y_pred_KNN)

sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu", fmt='g')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
labels = ['Iris Setosa','Iris Virginia','Iris Versicolor']
print(classification_report(Y_test, Y_pred_KNN))

```

```python
print(classification_report(Y_test, Y_pred_KNN))
```

## Decision Tree Classifier

```python
from sklearn import tree
classifier = tree.DecisionTreeClassifier()
DCT_Model = classifier.fit(X_train, Y_train)
Y_pred_DTC = DCT_Model.predict(X_test)
```

```python
accuracy_DTC = accuracy_score(Y_test, Y_pred_DTC)
print("Accuracy of the model is", accuracy_DTC*100)
cnf_matrix = confusion_matrix(Y_test, Y_pred_DTC)

sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu", fmt='g')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
labels = ['Iris Setosa','Iris Virginia','Iris Versicolor']
print(classification_report(Y_test, Y_pred_DTC))

```

# Conclusion
